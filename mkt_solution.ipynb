# E-commerce Marketing Performance & CRM Monetization Dashboard
* สร้างระบบรวบรวมข้อมูล (Data Pipeline) และแดชบอร์ดอัตโนมัติ เพื่อวิเคราะห์และตรวจสอบประสิทธิภาพการทำโฆษณา (Sponsored Solutions) ของร้านค้าบนแพลตฟอร์ม E-commerce พร้อมทั้งคำนวณรายได้ (Monetization) และจัดกลุ่มพฤติกรรมลูกค้า (CRM Analytics) เพื่อส่งต่ออินไซต์ให้ทีม Business และ Key Account Manager (KAM) นำไปใช้วางกลยุทธ์เพิ่มยอดขาย

### Data Simulation 

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# ตั้งค่า Seed เพื่อให้ได้ข้อมูลคงเดิมทุกครั้งที่รัน
np.random.seed(42)
n_rows = 1000

# 1. จำลองข้อมูลยอดขายและโฆษณา (Marketing Ads Performance)
start_date = datetime(2026, 1, 1)
dates = [start_date + timedelta(days=int(i)) for i in np.random.randint(0, 90, n_rows)]
merchant_ids = [f"MCH-{str(i).zfill(4)}" for i in np.random.randint(1, 101, n_rows)]
categories = np.random.choice(['Electronics', 'Fashion', 'FMCG', 'Home & Living'], n_rows)

impressions = np.random.randint(5000, 50000, n_rows)
clicks = (impressions * np.random.uniform(0.01, 0.05, n_rows)).astype(int)
cost = (clicks * np.random.uniform(2, 5, n_rows)).astype(int)
gmv_from_ads = (clicks * np.random.uniform(50, 200, n_rows)).astype(int)

df_ads = pd.DataFrame({
    'date': dates,
    'merchant_id': merchant_ids,
    'category': categories,
    'impressions': impressions,
    'clicks': clicks,
    'cost_thb': cost,
    'gmv_from_ads_thb': gmv_from_ads
})

# Export เป็นไฟล์ CSV ไว้ในเครื่อง Mac ของคุณ
df_ads.to_csv('marketing_ads_perf.csv', index=False)
print("สร้างไฟล์ marketing_ads_perf.csv สำเร็จ!")
```
